```python
import json
```

```python
with open('data.txt') as f:
    incidents = json.load(f)

with open('stateData.txt') as f:
    states = json.load(f)
```

```python
l = []
for i in states:
    l.append(i['state'])
    i['perCap2013'] = 0
    i['perCap2014'] = 0
    i['perCap2015'] = 0
    i['perCap2016'] = 0
    i['perCap2017'] = 0
    i['perCap2018'] = 0
    i['incidents2013'] = 0
    i['incidents2014'] = 0
    i['incidents2015'] = 0
    i['incidents2016'] = 0
    i['incidents2017'] = 0
    i['incidents2018'] = 0
    i['pop2013'] = int(i['pop2013'].strip('\n'))
    i['pop2014'] = int(i['pop2014'].strip('\n'))
    i['pop2015'] = int(i['pop2015'].strip('\n'))
    i['pop2016'] = int(i['pop2016'].strip('\n'))
    i['pop2017'] = int(i['pop2017'].strip('\n'))
    i['pop2018'] = int(i['pop2018'].strip('\n'))
```

```python
for i in incidents:
    year = i['date'][:4]
    state = i['state']
    state_index = l.index(state)
    states[state_index]['incidents'+year] += 1
```

```python
states
```

```python
for i in states:
    for j in range(2013, 2018):
        year = str(j)
        i['perCap'+year] = i['incidents'+year]/i['pop'+year]
```

```python
states
```

```python
with open("stateData.json", 'w') as out:
    out.write("[")
    for item in states:
        json.dump(item, out)
        out.write(",")
        out.write("\n")
    out.write("]")
```
