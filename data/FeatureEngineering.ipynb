```python
import json
```

```python
with open('data.txt') as f:
    incidents = json.load(f)

with open('stateData.txt') as f:
    states = json.load(f)
```

```python
l = []
for i in states:
    l.append(i['state'])
    i['perCap2013'] = 0
    i['perCap2014'] = 0
    i['perCap2015'] = 0
    i['perCap2016'] = 0
    i['perCap2017'] = 0
    i['perCap2018'] = 0
    i['incidents2013'] = 0
    i['incidents2014'] = 0
    i['incidents2015'] = 0
    i['incidents2016'] = 0
    i['incidents2017'] = 0
    i['incidents2018'] = 0
    i['injuries2013'] = 0
    i['injuries2014'] = 0
    i['injuries2015'] = 0
    i['injuries2016'] = 0
    i['injuries2017'] = 0
    i['injuries2018'] = 0
    i['deaths2013'] = 0
    i['deaths2014'] = 0
    i['deaths2015'] = 0
    i['deaths2016'] = 0
    i['deaths2017'] = 0
    i['deaths2018'] = 0
    i['pop2013'] = int(i['pop2013'].strip('\n'))
    i['pop2014'] = int(i['pop2014'].strip('\n'))
    i['pop2015'] = int(i['pop2015'].strip('\n'))
    i['pop2016'] = int(i['pop2016'].strip('\n'))
    i['pop2017'] = int(i['pop2017'].strip('\n'))
    i['pop2018'] = int(i['pop2018'].strip('\n'))
```

```python
for i in incidents:
    year = i['date'][:4]
    state = i['state']
    state_index = l.index(state)
    states[state_index]['incidents'+year] += 1
    states[state_index]['injuries'+year] += int(i['injuries'])
    states[state_index]['deaths'+year] += int(i['deaths'])
```

```python
for i in states:
    for j in range(2013, 2019):
        year = str(j)
        i['perCap'+year] = i['incidents'+year]/i['pop'+year]
```

```python
with open('stateData.json', 'w') as json_file:
    json.dump(states, json_file)
```

```python
distribution = dict()
for i in incidents:
    deaths = i['deaths']
    injuries = i['injuries']
    state = i['state']
    idd = i['id']
    year = i['date'][:4]
    if year in distribution:
        if state in distribution[year]:
            distribution[year][state].append({'deaths' : deaths, 'injuries' : injuries, 'id' : idd})
        else:
            distribution[year][state] = [{'deaths' : deaths, 'injuries' : injuries, 'id' : idd}]
    else:
        distribution[year] = {state : [{'deaths' : deaths, 'injuries' : deaths, 'id' : idd}]}
```

```python
with open('count_distribution.json', 'w') as json_file:
    json.dump(distribution, json_file)
```
